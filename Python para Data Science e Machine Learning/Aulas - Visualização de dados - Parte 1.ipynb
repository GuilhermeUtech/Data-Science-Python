<h1>Visualização de dados</h1>
<ul>
    <li>Matplotlib</li>
</ul>

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

```python
# Forma baseada em função (Funcional)
```

```python
plt.plot(x,y, color="blue") # O terceiro parâmetro define qual é a cor que você quer para o seu plot
plt.xlabel("Eixo X") # Define o texto que eu quero colocar no label do gráfico no eixo X
plt.ylabel("Eixo Y") # Define o texto que eu quero colocar no label do gráfico no eixo Y
plt.title("Gráfico de Curva") # Define o título do gráfico
# Como fazer o gráfico trastejado? plt.plot(x, y, "b--")
```

```python
# Caso eu não usa-se o ""%matplotlib inline" eu teria que usar plt.show() para mostrar o gráfico em tela
```

<h2>Plotando dois (ou mais) gráficos de uma vez</h2>

```python
plt.subplot(1, 2, 1) #Se eu bota-se nrows=3, ncols=3 eu teria um total de 9 gráficos, essa é a lógica. plt.subplot(nrows, ncols) 
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

<h3>Criando gráficos através da criação e manipulações das instancias de matplotlib </h3>

```python
fig = plt.figure()
```

```python
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
```

```python
fig = plt.figure() #Mesmo que já criado ali em cima tem que por aqui dentro para a visualização no notebook
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #Mesmo que já criado ali em cima tem que por aqui dentro para a visualização no notebook
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x,y)
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("Gráfico utilizando as instâncias de plt")
axes2.plot(y,x)
```
