<h1>Aula de visualização de dados</h1>

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

x = np.linspace(0, 5, 11)
y = x*x

fig, ax = plt.subplots() # Criando duas instâncias e já cria os eixos
ax.plot(x,x**2, 'r--') # Plot criado com orientação a objetos
ax.plot(x,x**3, 'b--') # Plot criado com orientação a objetos
ax.plot(x,x**4, 'g--') # Plot criado com orientação a objetos
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Título feliz")
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
```

```python
for axe in ax:
    axe.plot(x,y,'b')
    axe.set_title("Título do gráfico")
fig
```

<h2>Outra forma de fazer isso é acessar diretamente, afinal é um array</h2>

```python
ax[0].plot(x,y,'b')
ax[0].set_title("Título do gráfico")
```

```python
fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout() # Este método é utilizado para cuidar melhor dos espaçamentos, evitando sobreposição
```

```python
fig = plt.figure(figsize = (8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize = (5,8))
axes.plot(x, y, 'r--', label='Trabalhos Faculdade')
axes.plot(y,x,'b--', label='Meu tempo') #Esse parâmetro label é necessário para axes.legend()
axes.set_title("Título Feliz")
axes.legend()
#Dá para escolher a localização do label no gráfico, utilizando loc: axes.legend(loc= 0 [ou 1 ou 2 ou 3])
```
