```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

<h2>Vamos começar com o barplot</h2>
<p>Vai apresentar a média em barras dos dados passados como parâmetro</p>
<p>A linha em cima da barra representa o desvio padrão da barra</p>

```python
sns.barplot(x = 'sex', y = 'total_bill', data=tips)
```

```python
sns.barplot(x = 'sex', y = 'total_bill', data=tips, estimator=np.std) #passando funcao std por referência, vou ter o desvio padrão dos desvios padrões
```

<h2>Vamos para o countplot</h2>
<p>Vai ser uma simplificação do barplot onde o estimador vai ser um contador</p>


```python
sns.countplot(x = 'sex', data=tips)
```

<h2>Vamos apresentar agora o Boxplot</h2>
<p>Plota percentil de dados</p>


```python
sns.boxplot(x = 'day', y = 'total_bill', data=tips, palette="rainbow", hue='sex') #hue quebraria a apresentação dos dados em dois boxplots com a variação do dado passado como parâmetro
```

```python
sns.boxplot(data=tips, palette="rainbow", orient="h")
```

<h2>Vamos apresentar agora o violinplot</h2>
<p>Cria visualizações de dados por um modelo não paramétrico (KDE)</p>


```python
sns.violinplot(x = 'day', y = 'tip', data=tips, hue='sex', split=True)
```

<h2>Vamos apresentar agora o stripplot</h2>
<p></p>


```python
sns.stripplot(x = 'day', y = 'total_bill', data=tips, hue='sex', dodge='True')
```

<h2>Vamos apresentar agora o stripplot</h2>
<p></p>


```python
sns.swarmplot(x = 'day', y = 'total_bill', data=tips)
```

```python
sns.swarmplot(x = 'day', y = 'total_bill', data=tips, color='black')
sns.violinplot(x = 'day', y = 'total_bill', data=tips)
```

<h2>Vamos apresentar agora o factorplot</h2>
<p>Se liga nesse método</p>


```python
sns.factorplot(x = 'sex', y = 'total_bill', data=tips, kind='bar') #substituir o kind vai dizer qual método tu quer
```
