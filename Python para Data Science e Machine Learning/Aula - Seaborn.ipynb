<h2>Plots de distribuição</h2>
<ul>
    <li>distplt</li>
    <li>jointplt</li>

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips') #DataSet padrão do Seaborn|
```

```python
tips.head()
```

<h2>displot</h2>
<p>O que faz o <strong>displot</strong>? Ele pega teus dados e cria uma histograma desses dados e plota um gráfico de curva KDE</p>
   

```python
sns.distplot(tips['total_bill']) #Está ficando velho esse método e será retirado em versões futuras da biblioteca
sns.displot(tips['total_bill']) #Tipo o distplot mas não ploto a curva KDE
#sns.distplot(tips['total_bill'], kde=False) #Tirando o KDE
```

<h2>joinplot</h2>
<p>O que faz o <strong>jointplot</strong>? Ele cria uma curva de distruibuição semelhante a essa (displot) só que com duas variáveis ao mesmo tempo</p>
   

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind='reg') #kind='reg' é uma regressão linear, 'hex' faz um mapa de calor
```

```python
sns.jointplot(x = 'total_bill', y= 'tip', data = tips, kind='hex')
```

<h2>pairplot</h2>
<p>O que faz o <strong>pairplot</strong>? Pega todas as variáveis (numéricas) do dataset e faz um jointplot de cada uma delas</p>
   

```python
sns.pairplot(tips, hue='sex', palette='coolwarm') #cores diferentes para instancias diferentes do dataset, aqui temos os dados com o que cada sex representa
```

<h2>rugplot</h2>
<p>O que faz o <strong>rugplot</strong>? Não tão útil sozinho, mas ele é a base do KDE.</p>
   

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
sns.rugplot(tips['total_bill'])
```
