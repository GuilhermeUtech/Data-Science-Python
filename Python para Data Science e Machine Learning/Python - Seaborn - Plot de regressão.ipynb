<h1>Plots de regressão</h1>

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#Vamos fazer uma regressão entre o total da compra e a gorjeta
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day', aspect=0.6, height=8) #linear model plot
```
